```python
import pandas as pd
df1=pd.DataFrame(
    {
        'A': ['A0', 'A1', 'A2', 'A3'],
        'B': ['B4', 'B5', 'B6', 'B7'],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3'],
    },
    index=[0,1,2,3],
)
df1
```

```python
df2=pd.DataFrame(
    {
        'A': ['A4', 'A5', 'A6', 'A7'],
        'B': ['B4', 'B5', 'B6', 'B7'],
        'C': ['C4', 'C5', 'C6', 'C7'],
        'D': ['D4', 'D5', 'D6', 'D7'],
    },
    index=[0,1,2,3],
)
df2
```

```python
df3=pd.DataFrame(
    {
        'A': ['A8', 'A9', 'A10', 'A11'],
        'B': ['B8', 'B9', 'B10', 'B11'],
        'C': ['C8', 'C9', 'C10', 'C111'],
        'D': ['D8', 'D9', 'D10', 'D11'],
    },
    index=[8,9,10,11],
)
df3
```

```python
frames=[df1, df2, df3]
result=pd.concat(frames, keys=['X', 'Y', 'Z'], axis=1)
result
```

```python
df11=result.loc['Z']
df11
```

```python
df1.compare(df2, keep_shape=True, keep_equal=True)
```
